```
%matplotlib inline
#%matplotlib nbagg # JupyterLabでは上手く動かないので今回は使わない。
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from matplotlib.animation import ArtistAnimation
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

import sympy as sym
# Jupyter Notebook上で、レンダリングされた結果を表示する
sym.init_printing()
```

```
def return_ab2(x0,x1, y0,y1):
    a = (y1 - y0) / (x1 - x0)
    b = y0 -a*x0
   
    return (a, b)

def line2(x,a,b):
    y = a * x + b
    return y

def tangent_line2(i, x, y, x_tl):
    a, b = return_ab2(x[i], x[i+1], y[i], y[i+1])   
    y_tl = line2(x_tl, a, b)

    return y_tl

def here_true_diff(x):
    y = -2*x + 6

    return y
```

```
def here_function(x):
    y = -(x - 3) ** 2 + 9
    return y

start = 0 #定義域の左端
last = 6 # 定義域の右端
dx = 0.5

x0 = np.arange(start,last, 0.01)
y0 = here_function(x0)



fig = plt.figure(figsize=(12,4)) 
ax1 = fig.add_subplot(1,2,1) 
ax2 = fig.add_subplot(1,2,2) 
#ax.set_aspect('equal')

ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['bottom'].set_position('zero')
ax1.spines['left'].set_position('zero')
ax1.axis([-1,7.0,-0.0,10.0])
ax1.grid()
ax1.set_title('base y = -(x - 3)^2 + 9') # グラフタイトル
ax1.plot(x0, y0, label="base_function") # ラベルの追加
ax1.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10) # ラベルの反映ax1

ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['bottom'].set_position('zero')
ax2.spines['left'].set_position('zero')
ax2.grid() # グラフにグリッドを追加
ax2.set_title('diff dx:'+str(dx))
ax2.plot(x, y_diff, label='true_diff')
ax2.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10) # ラベルの反映ax2




x = np.arange(start,last+dx,dx)
y = here_function(x)


artist_list = [] # artistを入れていくリスト

x_limit0, x_limit1 = ax1.get_xlim()
ax1.plot(x0, y0) # 目標とする式

y_diff = here_true_diff(x)
point_list_x = []
point_list_y = []


for i in range(len(x)-1):
    x_tl = np.arange(x_limit0,x_limit1+dx,dx)
    y_tl = tangent_line2(i,x,y,x_tl) 
    art1 = ax1.plot(x_tl,y_tl, color='pink') # artistを作成

    a, b = return_ab2(x[i],x[i+1],y[i],y[i+1])
    point_list_x.append(x[i])
    point_list_y.append(a)
    art2 = ax2.plot(point_list_x,point_list_y, marker='.', color='red', markersize=4, linestyle='None') # 点

    artist_list.append(art1+art2)


ani = ArtistAnimation(fig, artist_list, interval = 200)
plt.close() 
HTML(ani.to_jshtml())
```
